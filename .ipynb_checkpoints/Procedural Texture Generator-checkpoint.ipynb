# Grid Noise texture: create nice stripy patterns

The grid noise widget

```python
from jupyter.ui_common import powerTwoSlider
from jupyter.make_ui import createUIforOP
createUIforOP(op= perlin.GridPerlin(), dWidgets = {"detail": powerTwoSlider(name = "Detail")})
```

Obtain a multiresolution texture

```python
details = [2**i for i in range(1,11)]
perlins = {"o"+str(i-1): perlin.GridPerlin(detail = details[i-1]) for i in range(1,11)}
coeffs = {"c" + str(i-1) : 2**(-i) for i  in range(1,11)}
```

```python
from operators.multiresolution import Multiresolution
perlins.update(coeffs)
mres = Multiresolution(**perlins)
mres.process()
```

```python
for i in range(10):
    mres["c" + str(i)] = 2**(-abs(3-i))
mres.process()
```

```python
from operators.threshold import Threshold

thres = Threshold(img = mres)
thres.process()
```

```python
thres['thres'] = 70
thres.process()
```

```python
from PIL.ImageOps import autocontrast, equalize
autocontrast(mres.process(), cutoff = 99)
```
